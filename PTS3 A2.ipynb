```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
dt= pd.read_csv("Assignment_2data.csv",header=None,sep=" ",decimal=",")

x=np.array(dt[0])
```

```python
def prob(x, theta=1): return x*np.exp(-x/theta)/theta**2

# def pdf_theta(x, theta=1,n=25): return np.exp(-0.5*(((x-theta)/((theta**2)/(2*n)))**2))/np.sqrt((2*np.pi*theta**2)/(2*n))
#this one is with the normal 

def pdf_theta_hat(x,theta=1,n=25): return x**(2*n-1)*np.exp(-x/(theta/(2*n)))/((theta/(2*n))**(2*n)*np.math.factorial(2*n-1))
```

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
y=np.linspace(0,5,10000)
plt.plot(y,pdf_theta_hat(y),"r");
plt.plot(y,pdf_theta_hat(y,theta=2),"g");
plt.plot(y,pdf_theta_hat(y,theta=3),"b");
```

```python
def Likelihood(theta): return np.prod(x*np.exp(-x/theta)/(theta**2))
theta=np.linspace(0,4,100)

plt.plot(theta,np.vectorize(Likelihood)(theta));
```

```python
def Log_L(theta): return np.log(Likelihood(theta))
plt.plot(theta,np.vectorize(Log_L)(theta))
plt.ylim((-100,-50));
```
