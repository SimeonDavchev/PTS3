```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
dt= pd.read_csv("Assignment_2data.csv",header=None,sep=" ",decimal=",")#this is the older version

x=np.array(dt[0])
```

```python
def prob(x, theta=1): return x*np.exp(-x/theta)/theta**2

def pdf_theta_hat(x,theta=1,n=25): return x**(2*n-1)*np.exp(-x/(theta/(2*n)))/((theta/(2*n))**(2*n)*np.math.factorial(2*n-1))
```

```python
#this will remove the warning messages for the plots
import warnings
warnings.filterwarnings("ignore")
```

```python
#2b
y=np.linspace(0,5,10000)
fig,ax=plt.subplots()
ax.plot(y,pdf_theta_hat(y),"r",label="1")
ax.plot(y,pdf_theta_hat(y,theta=2),"g",label="2")
ax.plot(y,pdf_theta_hat(y,theta=3),"b",label="3")
plt.title(r"Pdf of $\tilde{\theta}$")
plt.ylabel("Density"); plt.xlabel(r"$\tilde{\theta}$")
ax.legend(title=r"$\theta$");
```

```python
#2c
def Likelihood(theta): return np.prod(x*np.exp(-x/theta)/(theta**2))
theta=np.linspace(0,4,100)

plt.plot(theta,np.vectorize(Likelihood)(theta))
plt.ylabel(r"L($\theta$)");plt.xlabel(r"$\theta$")
plt.title("Likelihood Function");
```

```python
def Log_L(theta): return np.log(Likelihood(theta))
plt.plot(theta,np.vectorize(Log_L)(theta))
plt.ylim((-100,-50))
plt.ylabel(r"l($\theta$)"); plt.xlabel(r"$\theta$")
plt.title("Log Likelihood Function");
```
